```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft2,ifft2,fftshift
from speckle_fns import fits_import, circ_filter1
from speckle_fns import deconv0,deconv1, postprocess
%matplotlib inline
```

```python
# Import Reference and Double Stars, perform basic deconvolution
# Filenames
filePathDouble = "/home/niels/Desktop/KP330_PSD.fits"
filePathSingle = "/home/niels/Desktop/KP331_PSD.fits"
# Import FITS file data
fitsDoubleStar = fits_import(filePathDouble)
fitsSingleStar = fits_import(filePathSingle)
# Convert FITS Data to Floats
psdDoubleStar = fitsDoubleStar.astype(float)
psdSingleStar = fitsSingleStar.astype(float)
# Perform Deconvolution
constant = 1E-15
imgF = fftshift(deconv1(psdDoubleStar, psdSingleStar, constant))
img = postprocess(imgF)

# Display Deconvolution Image (Input to filter)
plt.figure(figsize = (16,6))
plt.subplot(1,2,1)
plt.imshow(np.log10(imgF))
plt.title("PSD of Deconvolved Image")
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(img)
plt.title("Autocorr of Deconvolved Image")
plt.colorbar()
```

```python
# Generating LPF

lpf_rad = 10

# Define gaussian function
def filter_fn_lpf(radius):
    return np.exp(-(np.power(radius,2)/(2*np.power(lpf_rad,2))))

# Calculate 2D Gaussian filter
lpfF = circ_filter1(radius = 254, size = 512, filter_fn = filter_fn_lpf)
lpf = np.abs(fftshift(ifft2(lpfF)))

# Displaying filter and its PSF
plt.figure(figsize = (16,6))
plt.subplot(1,2,1)
plt.imshow(np.log10(lpfF+1E-15))
plt.title("Filter in Freq Domain")
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(lpf)
plt.title("Filter in Spatial Domain")
plt.colorbar()
```

```python
# Generating HPF

hpf_rad = 6
hpf_amplitude = 0.8 

# Define LP gaussian function
def filter_fn_hpf(radius):  
    return np.multiply(hpf_amplitude,np.exp(-(np.power(radius,2)/(2*np.power(hpf_rad,2)))))

# Calculate 2D Gaussian filter
# Generate lowpass filter to serve as inverse of highpass filter
hpf = circ_filter1(radius = 254, size = 512, filter_fn = filter_fn_hpf)

# Generate array of 1's to be subtracted from
ones = np.zeros((512,512))
ones.fill(1)

# Calculate 1 - LP, which is HP
hpfF = np.subtract(ones,hpf)
hpf = np.abs(fftshift(ifft2(hpfF)))

# Displaying filter and its PSF
plt.figure(figsize = (16,6))
plt.subplot(1,2,1)
plt.imshow(np.log10(hpfF+1E-15))
plt.title("Filter in Freq Domain")
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(hpf)
plt.title("Filter in Spatial Domain")
plt.colorbar()
```

```python
# Interference Filter:

imgF_int = np.array(imgF)

# Perform filtering on vertical interference
for row in np.arange(512):
    good_pixels_avg = np.average(imgF[row,(255, 257)])
    imgF_int[row,256]=good_pixels_avg       
    
# Perform filtering on horizontal interference
for column in np.arange(512):
    good_pixels_avg = np.average(imgF[(255, 257),column])
    imgF_int[256,column]=good_pixels_avg   
    
    
img_int = postprocess(imgF_int)

# Displaying filter and its PSF
plt.figure(figsize = (16,6))
plt.subplot(1,2,1)
plt.imshow(np.log10(imgF_int+1E-15))
plt.title("With Interference Filter")
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(np.log10(imgF+1E-15))
plt.title("Without Interference Filter")
plt.colorbar()
```

```python
# Functions: 

# Filter Interference
def filter_interference(imgF):
   
    # Copy image to be filtered
    imgF_int = np.array(imgF)
    
    # Critical dimensions of image
    image_size = imgF_int.shape[0]
    int_index = image_size/2.0
    bad_pixels = np.array((int_index + 1, int_index - 1))

    # Perform filtering on vertical interference
    for row in np.arange(image_size):
        good_pixels_avg = np.average(imgF[row,(bad_pixels[0], bad_pixels[1])])
        imgF_int[row,256]=good_pixels_avg       
    
    # Perform filtering on horizontal interference
    for column in np.arange(512):
        good_pixels_avg = np.average(imgF[(bad_pixels[0], bad_pixels[1]),column])
        imgF_int[256,column]=good_pixels_avg  
        
    return imgF_int

# Perform LPF
def filter_lpf(imgF, lpf_rad):

    image_size = imgF.shape[0]
    circle_radius = (image_size/2)-2
    
    # Define gaussian function
    def filter_fn_lpf(radius):
        return np.exp(-(np.power(radius,2)/(2*np.power(lpf_rad,2))))

    # Calculate 2D Gaussian filter
    lpfF = circ_filter1(radius = circle_radius, size = image_size, filter_fn = filter_fn_lpf)
    
    # Return filtered fn
    return np.multiply(lpfF,imgF)

# Perform HPF
def filter_hpf(imgF, hpf_rad, hpf_amplitude):

    image_size = imgF.shape[0]
    circle_radius = (image_size/2)-2
    
    # Define LP gaussian function
    def filter_fn_hpf(radius):  
        return np.multiply(hpf_amplitude,np.exp(-(np.power(radius,2)/(2*np.power(hpf_rad,2)))))

    # Calculate 2D Gaussian filter
    # Generate lowpass filter to serve as inverse of highpass filter
    hpf = circ_filter1(radius = circle_radius, size = image_size, filter_fn = filter_fn_hpf)

    # Generate array of 1's to be subtracted from
    ones = np.zeros((image_size,image_size))
    ones.fill(1)

    # Calculate 1 - LP, which is HP
    hpfF = np.subtract(ones,hpf)
    
    # Return Filtered
    return np.multiply(hpfF,imgF)
```

```python
# Test Filters

# Apply Interference Filter
imgF_filtered = filter_interference(imgF)
plt.figure(figsize = (20,6))
plt.subplot(1,3,1)
plt.imshow(np.log10(imgF_filtered+1E-15))
plt.title("With Interference Filter")
plt.colorbar()

# Apply LPF
imgF_filtered = filter_lpf(imgF_filtered, 10)
plt.subplot(1,3,2)
plt.imshow(np.log10(imgF_filtered+1E-15))
plt.title("With Interference Filter + LPF")
plt.colorbar()

# Apply HPF
imgF_filtered = filter_hpf(imgF_filtered, 6, 0.8)
plt.subplot(1,3,3)
plt.imshow(np.log10(imgF_filtered+1E-15))
plt.title("With Interference Filter + LPF + HPF")
plt.colorbar()

# Postprocess
img_autocorr = postprocess(imgF_filtered)
plt.figure(figsize = (6,6))
plt.imshow(img_autocorr)
plt.title("Autocorr of Filtered Image")
plt.colorbar()
```
